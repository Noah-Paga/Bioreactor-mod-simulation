### Derivation of the Lineweaver-Burk Equation

The derivation begins with the Michaelis-Menten equation:

$$v = \frac{v_{max}[S]}{K_m + [S]} \tag{1}$$

To linearize this relation, we take the reciprocal of both sides:

$$\frac{1}{v} = \frac{K_m + [S]}{V_{max}[S]} \tag{2}$$

Next, by distributing the denominator across the terms in the numerator to separate the variables we arrive at the Lineweaver-Burk format:

$$\frac{1}{v} = \frac{K_m}{V_{max}} \cdot \frac{1}{[S]} + \frac{1}{V_{max}} \tag{3}$$

The key parameters from the Lineweaver-Burk plot are:

$$\text{y-intercept} = \frac{1}{V_{max}} \quad \left(\text{when } \frac{1}{[S]} = 0\right)$$

$$\text{x-intercept} = -\frac{1}{K_m} \quad \left(\text{when } \frac{1}{v} = 0\right)$$

$$\text{Slope} = \frac{K_m}{V_{max}} \tag{4}$$

### (a) Conversion to Undissociated Acid Concentration

The inhibitory effect of lactic acid on *S. cremoris* is caused only by the undissociated form of the acid (HA), since it is the only form able to cross the cell membrane and disturb the internal pH of the cell. $$f_{HA} = \frac{1}{1 + 10^{\text{pH} - pK_a}} \tag{6}$$

$$P\,[\text{mM}] = f_{HA} \times \frac{p_{\text{total}}\,[\text{g}\,\text{L}^{-1}]}{M_w\,[\text{g}\,\text{mol}^{-1}]} \times 1000 \tag{7}$$

### Part (a): Mass Balance

The reversible reaction given:

$$E + S \underset{k_{1,b}}{\overset{k_{1,f}}{\rightleftharpoons}} ES \underset{k_{2,b}}{\overset{k_{2,f}}{\rightleftharpoons}} E + P \tag{11}$$

The general mass balance for a species $i$ in a reactor is given by:

$$\frac{dN_i}{dt} = \dot{N}_{i,in} - \dot{N}_{i,out} + \dot{N}_{i,gen} \tag{12}$$

For a batch reactor, there is no inlet or outlet flow. Therefore the balance simplifies to:

$$\frac{dN_i}{dt} = \dot{N}_{i,gen} \tag{13}$$

Since the reaction takes place in a reactor with constant volume $V$, we can write:

$$N_i = C_i V \tag{14}$$

which leads to the following expression where $r_i$ is the net rate of formation of species $i$:

$$\frac{dC_i}{dt} = r_i \tag{15}$$

For the reversible reaction given, the elementary reaction rates are:

$$v_{1,f} = k_{1,f}[E][S] \quad v_{1,b} = k_{1,b}[ES] \quad v_{2,f} = k_{2,f}[ES] \quad v_{2,b} = k_{2,b}[E][P] \tag{16}$$

For the substrate $S$:

$$\frac{d[S]}{dt} = -v_{1,f} + v_{1,b} \tag{17}$$

$$\frac{d[S]}{dt} = -k_{1,f}[E][S] + k_{1,b}[ES] \tag{18}$$

Following the same procedure, the mass balance for the enzyme–substrate $ES$, the product $P$ and the free enzyme $E$ is obtained:

$$\frac{d[ES]}{dt} = k_{1,f}[E][S] - k_{1,b}[ES] - k_{2,f}[ES] + k_{2,b}[E][P] \tag{19}$$

$$\frac{d[P]}{dt} = k_{2,f}[ES] - k_{2,b}[E][P] \tag{20}$$

$$\frac{d[E]}{dt} = -k_{1,f}[E][S] + k_{1,b}[ES] + k_{2,f}[ES] - k_{2,b}[E][P] \tag{21}$$

</div>

<div style='background-color: white; color: black;'>

### Part (b): QSSA approximation and expression of [ES]

In order to determine the concentration of the enzyme and substrate complex, the quasi–steady state approximation was applied to the mass balance of $ES$, assuming that $\frac{d[ES]}{dt} = 0$. By rearranging the mass balance equation and grouping the terms containing $[E]$ and $[ES]$, the following relation between the free enzyme concentration and the complex concentration is obtained:

$$(k_{1,f}[S] + k_{2,b}[P])[E] = (k_{1,b} + k_{2,f})[ES] \tag{22}$$

The total enzyme concentration is conserved such that $[E]_0 = [E] + [ES]$, which allows the free enzyme concentration to be written as $[E] = [E]_0 - [ES]$. Substituting this expression into the previous equation eliminates the free enzyme concentration. After rearranging the resulting expression and isolating the enzyme–substrate complex concentration, the following expression for $[ES]$ was obtained:

$$[ES] = \frac{[E]_0\left(k_{1,f}[S] + k_{2,b}[P]\right)}{k_{1,b} + k_{2,f} + k_{1,f}[S] + k_{2,b}[P]} \tag{23}$$

From the mass balance of the product $[P]$ as shown in Eq. (20) and substituting the expressions of $[ES]$ and $[E] = [E]_0 - [ES]$, the rate of formation of $[P]$ is given by:

$$\frac{d[P]}{dt} = \frac{[E]_0\left(k_{1,f}k_{2,f}[S] - k_{1,b}k_{2,b}[P]\right)}{k_{1,b} + k_{2,f} + k_{1,f}[S] + k_{2,b}[P]} \tag{24}$$

</div>

<div style='background-color: white; color: black;'>

### Part (c): Reversible Michaelis–Menten rate expression

Using the rate expression of the product derived in part b) shown in Eq. (24), the denominator can be factorized by $(k_{1,b} + k_{2,f})$ in order to rewrite the equation in the reversible Michaelis–Menten form. This gives:

$$K_{m,S} = \frac{k_{1,b} + k_{2,f}}{k_{1,f}} \qquad K_{m,P} = \frac{k_{1,b} + k_{2,f}}{k_{2,b}} \qquad V_{max,f} = [E]_0 k_{2,f} \qquad V_{max,b} = [E]_0 k_{1,b} \tag{25}$$

Therefore:

$$v([S],[P]) = \frac{\dfrac{V_{max,f}[S]}{K_{m,S}} - \dfrac{V_{max,b}[P]}{K_{m,P}}}{1 + \dfrac{[S]}{K_{m,S}} + \dfrac{[P]}{K_{m,P}}} \tag{26}$$

From the four parameters $V_{max,f}$, $V_{max,b}$, $K_{m,S}$ and $K_{m,P}$ alone, the four elementary rate constants cannot be determined because $V_{max,f}$ and $V_{max,b}$ depend on the products $[E]_0 k_{2,f}$ and $[E]_0 k_{1,b}$.

If the total enzyme concentration $[E]_0$ is known, the elementary rate constants can be written as:

$$k_{2,f} = \frac{V_{max,f}}{[E]_0} \qquad k_{1,b} = \frac{V_{max,b}}{[E]_0} \qquad k_{1,f} = \frac{k_{1,b} + k_{2,f}}{K_{m,S}} \qquad k_{2,b} = \frac{k_{1,b} + k_{2,f}}{K_{m,P}} \tag{27}$$

</div>

<div style='background-color: white; color: black;'>

### Part (d): Determination of kinetic constants and ODE simulation

Using the expression of $K_{m,S}$ and $K_{m,P}$ derived in Question c, the missing elementary rate constants can be determined from the given data:

$$k_{1,b} = 10\,\text{s}^{-1} \qquad k_{2,f} = 22\,\text{s}^{-1} \qquad K_{m,S} = 210\,\mu\text{M} \qquad K_{m,P} = 97\,\mu\text{M} \tag{28}$$

Thus, the four elementary rate constants obtained are:

$$k_{1,b} = 10\,\text{s}^{-1} \qquad k_{2,f} = 22\,\text{s}^{-1} \qquad k_{1,f} = 1.52 \times 10^5\,\text{L}\,\text{mol}^{-1}\text{s}^{-1} \qquad k_{2,b} = 3.30 \times 10^5\,\text{L}\,\text{mol}^{-1}\text{s}^{-1} \tag{29}$$

From the enzyme–substrate concentration derived in Question b (23) and using the enzyme conservation $[E]_0 = [E] + [ES]$, the mass balances of the substrate (18) and the product (20) can be rewritten as:

$$\frac{d[S]}{dt} = -k_{1,f}([E]_0 - [ES])[S] + k_{1,b}[ES] \tag{30}$$

$$\frac{d[P]}{dt} = k_{2,f}[ES] - k_{2,b}([E]_0 - [ES])[P] \tag{31}$$

The system is solved numerically on the interval $t \in [0, 5]\,\text{s}$ using the initial conditions:

$$[S_0] = 1\,\text{mol}\,\text{L}^{-1} \qquad [E_0] = 0.1\,\text{mol}\,\text{L}^{-1} \qquad [P_0] = 0\,\text{mol}\,\text{L}^{-1} \tag{32}$$

As shown in the plot 7a, the substrate concentration decreases rapidly while the product concentration increases due to the enzymatic conversion. As time progresses, the system approaches equilibrium and both concentrations stabilize around $0.5\,\text{mol}\,\text{L}^{-1}$. This reflects the reversibility of the reaction and the formation of an equilibrium during the reaction.

### System Overview

Steady-state chemostat data were collected for yeast growth on a carbon substrate with feed concentration $S_0 = 10\,\text{g/L}$ and excess dissolved oxygen. All modelling assumes Monod kinetics, a constant yield coefficient $Y_{X/S}$, and a linear maintenance term $m$ on substrate consumption:

### Part (a): Estimation of $\mu_{max}$ and $K_s$

At steady state the chemostat imposes $\mu = D$, so the Monod equation becomes:

$$D = \frac{\mu_{max} S}{K_s + S}$$

The seven pre-washout data points ($D \leq 0.80\,\text{h}^{-1}$) were used for parameter estimation; the washout point ($D = 0.84\,\text{h}^{-1}$) is excluded because the culture is no longer at a viable steady state. Parameters were estimated using the Lineweaver-Burk linearization, obtained by taking the reciprocal of the Monod equation:

$$\frac{1}{D} = \frac{K_s}{\mu_{max}} \cdot \frac{1}{S} + \frac{1}{\mu_{max}}$$

A linear regression of $1/D$ against $1/S$ yields $\mu_{max}$ from the $y$-intercept and $K_s$ from the slope, giving:

$$\mu_{max} = 2.298\,\text{h}^{-1}, \qquad K_s = 0.548\,\text{g/L}$$

### Part (b): Estimation of $Y_{X/S}$ and Maintenance Coefficient $m$

### Derivation of the Modified Lineweaver-Burk Equation 5

The derivation of the Michaelis-Menten (MM) equation in presence of a competitive inhibitor I (modified MM equation) is shown below.

The reaction mechanism is as follows:
- **Substrate Binding:** $E + S \rightleftharpoons ES \xrightarrow{k_{cat}} E + P$
- **Inhibitor Binding:** $E + I \rightleftharpoons EI$

The dissociation constant for the inhibitor $K_I$ is defined as:

$$K_I = \frac{[E][I]}{[EI]} \rightarrow [EI] = \frac{[E][I]}{K_I} \tag{37}$$

Now we use the mass balance for the enzyme (conservation of the enzyme), which in this modified situation is given by:

$$[E]_0 = [E] + [ES] + [EI] \tag{38}$$

Substituting the expression for $[EI]$ from Equation 37 into the mass balance:

$$[E]_0 = [E] + [ES] + \frac{[E][I]}{K_I} = [E]\left(1 + \frac{[I]}{K_I}\right) + [ES] \tag{39}$$

By defining the degree of inhibition as $\alpha = 1 + \frac{[I]}{K_I}$, we can isolate the enzyme $[E]$ and find:

$$[E]_0 = [E]\alpha + [ES] \Rightarrow [E] = \frac{[E]_0 - [ES]}{\alpha} \tag{40}$$

Next, by substituting the expression for E derived in Equation 40 into the standard definition of the Michaelis constant, $K_m = \frac{[E][S]}{[ES]}$:

$$K_m = \frac{\left(\frac{[E]_0 - [ES]}{\alpha}\right)[S]}{[ES]} \tag{41}$$

The concentration of the enzyme-substrate complex can now be found:

$$[ES] = \frac{[E]_0[S]}{K_m\alpha + [S]} \tag{42}$$

The reaction rate is defined as $v = k_{cat}[ES]$ and maximum velocity as $V_{max} = k_{cat}[E]_0$. Multiplying Equation 42 by $k_{cat}$ yields the modified Michaelis-Menten equation:

$$v = \frac{V_{max}[S]}{K_m\alpha + [S]} \tag{43}$$

Substituting the definition of $\alpha$ back into the equation gives the final form for competitive inhibition:

$$v = \frac{V_{max}[S]}{K_m\left(1 + \frac{[I]}{K_I}\right) + [S]} \tag{44}$$

Finally to derive the Lineweaver-Burk equation 5 shown in Exercise 1, part d, we take the reciprocal of both sides of the modified MM equation:

$$\frac{1}{v} = \underbrace{\frac{K_m}{V_{max}}\left(1 + \frac{[I]}{K_I}\right)}_{\text{Slope}} \cdot \frac{1}{[S]} + \underbrace{\frac{1}{V_{max}}}_{\text{y-intercept}} \tag{45}$$

### Exercise 2

#### Conversion to Undissociated Acid Concentration

The inhibitory effect of lactic acid is due to its undissociated form (HA), which is able to diffuse through the cell membrane. The relationship between the dissociated and undissociated forms is described by the Henderson–Hasselbalch equation:

$$\text{pH} = pK_a + \log\left(\frac{[A^-]}{[HA]}\right) \tag{46}$$

Rearranging this equation gives:

$$\frac{[A^-]}{[HA]} = 10^{\text{pH} - pK_a} \tag{47}$$

The fraction of undissociated acid is defined as:

$$f_{HA} = \frac{[HA]}{[HA] + [A^-]} \tag{48}$$

Substituting the ratio yields:

$$f_{HA} = \frac{1}{1 + 10^{\text{pH} - pK_a}} \tag{49}$$

The undissociated acid concentration $P$ (in mM) is obtained from the total lactic acid concentration:

$$P = f_{HA} \cdot \frac{p_{\text{total}}}{M_w} \cdot 1000 \tag{50}$$

where $M_w$ is the molar mass of lactic acid.

#### Model 1: Hyperbolic Inhibition

The inhibition model is expressed as:

$$\mu = \frac{\mu_{max}}{1 + P/K_I} \tag{51}$$

This equation is nonlinear with respect to the parameter $K_I$. To allow the use of linear regression, it is transformed by taking the reciprocal:

$$\frac{1}{\mu} = \frac{1}{\mu_{max}} + \frac{P}{\mu_{max} K_I} \tag{52}$$

This expression has the form of a linear equation $y = a + bx$ with:

$$y = \frac{1}{\mu}, \quad x = P, \quad a = \frac{1}{\mu_{max}}, \quad b = \frac{1}{\mu_{max} K_I} \tag{54}$$

A linear regression is therefore performed on $1/\mu$ as a function of $P$ using polyfit. Once the slope $b$ is obtained, the inhibition constant is calculated as:

$$K_I = \frac{1}{b\,\mu_{max}} \tag{55}$$

It should be noted that this linearisation modifies the distribution of experimental errors, since the regression is performed on $1/\mu$ instead of $\mu$.

#### Model 2: Linear Inhibition

The second model assumes a linear decrease of the growth rate with inhibitor concentration:

$$\mu = \mu_{max}\left(1 - \frac{P}{P_{max}}\right) \tag{56}$$

In this case, $P_{max}$ represents the concentration at which growth stops ($\mu = 0$). This model is directly linear in $\mu$, which allows parameter estimation without any transformation.

#### Effect of pH on Growth Rate

The undissociated acid concentration depends on pH according to:

$$P = \frac{p_{\text{total}}}{M_w \cdot 1000} \cdot \frac{1}{1 + 10^{\text{pH} - pK_a}} \tag{57}$$

Growth stops when $\mu = 0$, which corresponds to $P = P_{max}$. Substituting and solving for pH gives:

$$\text{pH} = pK_a + \log\left(\frac{p_{\text{total}}}{M_w \cdot 1000 \cdot P_{max}} - 1\right) \tag{59}$$

This expression allows the determination of the critical pH below which growth ceases for a given total lactic acid concentration.

### Exercise 4

#### Part (c): Derivation of $D_{max}$ and $D_{opt}$

**Derivation of $D_{max}$**

At steady state, the chemostat biomass balance with sterile feed ($X_0 = 0$) gives $\mu = D$. Substituting the Monod equation and solving for the steady-state substrate concentration $S$:

$$D(K_s + S) = \mu_{max} S \Rightarrow D K_s = (\mu_{max} - D) S \Rightarrow S(D) = \frac{D K_s}{\mu_{max} - D}$$

As $D$ increases, $S$ increases monotonically. Washout occurs when the substrate concentration inside the reactor reaches the feed concentration ($S \rightarrow S_0$), meaning no substrate is being consumed because there is no viable biomass left. Setting $S = S_0$:

$$S_0 = \frac{D_{max} K_s}{\mu_{max} - D_{max}} \Rightarrow S_0(\mu_{max} - D_{max}) = D_{max} K_s \Rightarrow S_0 \mu_{max} = D_{max}(K_s + S_0)$$

$$D_{max} = \frac{\mu_{max} S_0}{K_s + S_0} = 2.1787\,\text{h}^{-1}$$

**Derivation of $D_{opt}$**

**Step 1:** Write $X(D)$ explicitly (ignoring maintenance, $m = 0$). The steady-state substrate balance simplifies to:

$$D(S_0 - S) = \frac{DX}{Y_{X/S}} \Rightarrow X = Y_{X/S}(S_0 - S)$$

Substituting $S(D)$ from above:

$$X(D) = Y_{X/S}\left(S_0 - \frac{D K_s}{\mu_{max} - D}\right) = Y_{X/S} \cdot \frac{\mu_{max} S_0 - D(K_s + S_0)}{\mu_{max} - D}$$

**Step 2:** Form the productivity $P = D \cdot X$. Let $A = \mu_{max} S_0$ and $B = K_s + S_0$ for brevity:

$$P(D) = D \cdot X(D) = Y_{X/S} \cdot \frac{D(A - BD)}{\mu_{max} - D}$$

**Step 3:** Differentiate and set to zero. Applying the quotient rule and setting $f'(D) = 0$ requires the numerator to vanish:

$$(A - 2BD)(\mu_{max} - D) + D(A - BD) = 0$$

Expanding and simplifying:

$$A\mu_{max} - 2BD\mu_{max} + BD^2 = 0$$

Dividing through by $B$:

$$D^2 - 2\mu_{max} D + \frac{A}{B}\mu_{max} = 0$$

**Step 4:** Solve the quadratic. Recalling $A/B = \mu_{max} S_0/(K_s + S_0)$:

$$D = \mu_{max}\left(1 \pm \sqrt{\frac{K_s}{K_s + S_0}}\right)$$

The $+$ root exceeds $D_{max}$ and is therefore non-physical; taking the $-$ root:

$$D_{opt} = \mu_{max}\left(1 - \sqrt{\frac{K_s}{K_s + S_0}}\right) = 1.7724\,\text{h}^{-1}$$

Both $D_{max}$ and $D_{opt}$ are well below $\mu_{max}$ because $K_s$ and $S_0$ are of comparable magnitude here, meaning the washout boundary is not approached as closely as in systems where $S_0 \gg K_s$.
