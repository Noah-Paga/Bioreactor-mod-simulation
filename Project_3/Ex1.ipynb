Plotting the cell concentrations and productivities as a function of the dilution rate for different values of 𝛼. 

Below are the plots of the selectivity for $\alpha$ values: 0, 0.3, 0.6, and 0.8.

```python
import numpy as np
import matplotlib.pyplot as plt

# Values given in the exercise
mu_m = 1.0      # Max specific growth rate
Ks = 0.01       # Half-saturation constant
Yxs = 0.5       # Yield 
S0 = 2.0        # Feed substrate concentration
C = 2.0         # Concentration factor

# Recycle ratios (alpha's) chosen to compare
alphas = [0, 0.3, 0.6, 0.8]
colors = ['black', 'green', 'orange', 'red']

plt.figure(figsize=(12, 8))

for alpha, color in zip(alphas, colors):
    D_max = mu_m / (1 - alpha)
    # Dilution rates generated from 0.05 to just below D_max
    D = np.linspace(0.05, D_max - 0.02, 200)
    mu = D * (1 + alpha - alpha * C)
    S = (Ks * mu) / (mu_m - mu)
    X = (Yxs * (S0 - S)) / (1 + alpha - alpha * C)
    X2 = X * (1 + alpha - alpha * C)
    Productivity = D * X2

    plt.subplot(2, 1, 1)
    plt.plot(D, X, label=fr'$\alpha$ = {alpha}', color=color, linewidth=2)
    plt.subplot(2, 1, 2)
    plt.plot(D, Productivity, label=fr'$\alpha$ = {alpha}', color=color, linewidth=2)

plt.subplot(2, 1, 1)
plt.title('Cell Concentration (X) inside Reactor vs. Dilution Rate', fontsize=14)
plt.ylabel('X (g/L)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.subplot(2, 1, 2)
plt.title('Volumetric Productivity (D * X2) vs. Dilution Rate', fontsize=14)
plt.xlabel('Dilution Rate (D, h^-1)', fontsize=12)
plt.ylabel('Productivity (g/L/h)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.tight_layout()
plt.show()
```
