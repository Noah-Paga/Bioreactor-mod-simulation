# Exercice 1:

Dans le PDF les équations exactes des courbes (étiquetées (1) et (2)) sont insérées comme images et n'ont pas été extraites automatiquement. Ci-dessous un script complet qui crée une figure 2×2, applique les options demandées (log, 3D surface, style, légendes, etc.) et laisse en clair où coller les formules exactes du PDF. Lancez-le tel quel (nécessite numpy, matplotlib et mpl_toolkits.mplot3d).

```python
# ex1_plot.py
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D   # noqa: F401
```

```python
# x vector for (a)-(c)
x = np.arange(0, 5+0.1, 0.1)

# --- DEFINE y-curves ---
# REMARQUE: dans votre PDF les équations (1) sont en image.
# Remplacez y_a, y_b, y_c ci-dessous par les formules exactes du PDF.
y_a = np.sin(x)               # <-- exemple : remplacer par la formule (1a)
y_a2 = np.cos(2*x)            # <-- exemple : seconde courbe (pour (a))
y_b_x = np.arange(0, 5+0.5, 0.5)   # points discrets pour (b)
y_b = np.exp(-y_b_x)               # <-- remplacer
y_c = np.exp(-x)                   # <-- remplacer, (y-axis log)

# --- subplot (d): surface z(x,y) ---
X = np.arange(-10, 10+0.25, 0.25)
Y = np.arange(-10, 10+0.25, 0.25)
XX, YY = np.meshgrid(X, Y)
# Remplacer la formule ci-dessous par l'équation (2) du PDF
Z = np.sin(np.sqrt(XX**2 + YY**2)) / (1 + 0.1*(XX**2 + YY**2))

fig = plt.figure(figsize=(12, 9))

# (a)
ax1 = plt.subplot(2,2,1)
ax1.plot(x, y_a, linewidth=6, label='courbe 1')
ax1.plot(x, y_a2, linewidth=2, label='courbe 2')
ax1.set_title('(a) Deux courbes, légende déplacée', fontsize=12)
ax1.legend(loc='upper right', fontsize=10)
ax1.grid(True)

# (b)
ax2 = plt.subplot(2,2,2)
ax2.plot(y_b_x, y_b, 'x', markersize=8)   # marqueurs (cross)
ax2.set_xlim([0, 5])
ax2.set_ylim([min(y_b)-0.1, max(y_b)+0.1])
ax2.set_title('(b) points discrets', fontsize=12)
ax2.grid(True)

# (c)
ax3 = plt.subplot(2,2,3)
ax3.semilogy(x, np.abs(y_c))   # y-axis logarithmique
ax3.set_title('(c) échelle logarithmique', fontsize=12)
ax3.text(0.5, 0.5, 'y = ... (mettre l\'équation ici)', transform=ax3.transAxes)
ax3.grid(True, which='both')

# (d) 3D surface
ax4 = plt.subplot(2,2,4, projection='3d')
ax4.plot_surface(XX, YY, Z, rstride=1, cstride=1, linewidth=0, antialiased=True)
ax4.set_title('(d) surface 3D', fontsize=12)

plt.tight_layout()
plt.show()
```

# Exercice 2:

```python
# données
t_min = np.array([4,20,48,76,120], dtype=float)
V_L   = np.array([115,365,680,850,1130], dtype=float)

# conversions
t_s = t_min * 60.0        # s
V_m3 = V_L * 1e-3         # m^3

t_over_V = t_s / V_m3

# régression linéaire t/V = m * V + b
m, b = np.polyfit(V_m3, t_over_V, 1)

# paramètres donnés
mu = 2.9e-3   # Pa.s
a  = 4.0      # m/kg
C  = 1920.0   # kg/m^3
A  = 0.28     # m^2

DeltaP = mu * a * C / (2 * A**2 * m)
Rm = b * A * DeltaP / mu

print("m (s/m^6) =", m)
print("b (s/m^3) =", b)
print("DeltaP (Pa) =", DeltaP)
print("Rm (1/m) =", Rm)

# optionnel : tracer t/V vs V et droite fit
import matplotlib.pyplot as plt
V_plot = np.linspace(min(V_m3), max(V_m3), 100)
plt.plot(V_m3, t_over_V, 'o', label='données')
plt.plot(V_plot, m*V_plot + b, '-', label='fit linéaire')
plt.xlabel('V (m^3)')
plt.ylabel('t/V (s/m^3)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
y0 = [5,0,0,0,1,0]
p = {'k1':5,'k2':1,'k3':30,'k4':2,'k5':3}
def rhs(t,y):
    A,B,C,D,X,Y = y
    k1,k2,k3,k4,k5 = p['k1'],p['k2'],p['k3'],p['k4'],p['k5']
    r1 = k1*A*X
    r2 = k2*B
    r3 = k3*C*Y
    r4 = k4*D
    r5 = k5*B
    return [-r1, r1-r2-r5+r4, r2-r3, r5+r3-r4, -r1+r3, r1-r3]

sol = solve_ivp(rhs, [0,30], y0, t_eval=[30])
print("Concs at t=30 h: ", sol.y[:, -1])
```
